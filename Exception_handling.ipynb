```python
# inheritance

class A:
    my_var = 34
    def a_method(self):
        return "i am in class A (object)"

class B(A):                                  # B(A) means B object can now accesses class A method or element.
    var = 200
    def b_method(self):
        return "i am in class B (object)"

b_object = B()
b_object.a_method()
```

```python
b_object.my_var
```

```python

def div(num1:int, num2:int) -> float:
    if not isinstance(num1,(int,float)) or not isinstance(num2,(int,float)):
        raise TypeError("num1 and num2 should be folat or int only")
    if num2 == 0:
        raise ValueError("num2 should not be zero")
    return num1/num2
```

```python
try:
    a =  45
    b = 0
    print(div(a,b))
except ZeroDivisionError as e:
    print(e)
except Exception as e:
    print( e)

print("Hello")
```

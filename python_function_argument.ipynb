```python
#  types of arguments
# positional args
# keyword args
# position-only (/)
# keyword-only (*)
# default 
# variable-length positional  (*args  arbitrary args)
# variable-length keywords  (**kwargs  arbitrary keywords args)
```

```python
def greatting(fname, lname):
    '''dock string create : this function is use to greatting by entering fname and lname'''
    print(f" Hii {fname} {lname} how are you!")
```

```python
greatting("roshan","kumar")       # positional args
```

```python
greatting("kumar","roshan")
```

```python
greatting( lname = "kumar", fname = "roshan")      #  keyword args
```

```python
def greatting(fname, lname,/):   # positional-only
    '''dock string create : this function is use to greatting by entering fname and lname'''
    print(f" Hii {fname} {lname} how are you!")
```

```python
greatting("roshan","kumar")       # positional args
```

```python
greatting( lname = "kumar", fname = "roshan")      #  keyword args
```

```python
def greatting(*,fname, lname):   # keyword-only
    '''dock string create : this function is use to greatting by entering fname and lname'''
    print(f" Hii {fname} {lname} how are you!")
```

```python
greatting("roshan","kumar")       # positional args
```

```python
greatting( lname = "kumar", fname = "roshan")      #  keyword args
```

```python
# for loop
# iterable item (iteration: no. of time the loop run known as iteration)
# list
# tuple
# set
# dict
# str
# range()
# array
# namedtuple
# frozenset
```

```python
list1 = [1,4,3,6,11,43,"70",76,90,"70"]
```

```python
count = 0
for item in list1:
    if type(item) == str:
        print("string item:", item, end ='')
        print(" ;string index:", count)

    count = count + 1
```

```python
for i in 1234:
    print(i)   # int is not iterable
```

```python
for i in range(11):   # range(11) == range(0,11,1) ;0 is start; 11 is stop; 1 is step
    print(i)
```

```python
for i in range(1,12,2):
    print(i)
```

```python
# printing tables using for loop
n = int(input("Enter num:"))

for i in range(1,11,1):
    print(f"{n} x {i} = {n*i}")
```

```python
# loop doesn't stop by it self until all element not covered.
list1 = ["70",4,3,6,11,43,76,90,]

for i in list1:
    if i == "70":                       # we found at  1st place.
        print(f"item is found: {i} ")
    else:
        print("not found:")              # so the rest of loop is just waste of time and space.
```

```python
# Break
# break is used to immediately stop a loop (for or while), even if the loop hasn’t finished.

list1 = ["70",4,3,6,11,43,76,90,]

for i in list1:
    if i == "70":                       # we found at  1st place.
        print(f"item is found: {i} ")
        break
    else:
        print("not found:") 
```

```python
# Continue
# continue is used to skip the current iteration of a loop and move to the next iteration.

list1 = [1,3,4,"70",4,3,6,11,43,76,90,]
total = 0

for i in list1:
    if type(i) != int:
        continue                # continue skip the string iteration
    total = total + i 
        
print("Sum of element:", total) 
```

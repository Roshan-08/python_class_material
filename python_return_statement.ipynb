```python
# function can return multiple values
def operation(num1:int, num2:int) -> tuple:
    addition = num1 + num2
    multi = num1 * num2

    return addition, multi
```

```python
result = operation(12,11)
```

```python
# using *args
def addition(*num):
    print(type(num))
    return sum(num)
```

```python
addition(12,14)
```

```python
addition(11,21,41,33,15)    # by using *args we can  add as much  value we want.
```
